# **INTRODUCCION**


El Mantenimiento Predictivo (MPD) es una estrategia clave en la gestión industrial que busca anticipar fallos en equipos mediante el análisis de datos operativos. A diferencia del mantenimiento preventivo, que se basa en intervenciones programadas, el MPD utiliza datos en tiempo real para estimar con precisión el momento en que ocurrirá una falla, permitiendo una intervención proactiva. Esto no solo minimiza costos operativos y tiempos de inactividad, sino que también optimiza la seguridad y eficiencia en entornos industriales.

En la última década, el avance del Internet de las Cosas (IoT) ha facilitado la implementación del MPD mediante la recopilación masiva de datos a través de sensores en equipos rotativos y sistemas eléctricos. Variables como vibración, temperatura, voltaje y presión pueden monitorearse continuamente, proporcionando información clave para modelos de predicción. Entre las metodologías más avanzadas para este análisis, el Análisis de Supervivencia y las técnicas de Machine Learning han demostrado ser altamente efectivas para estimar la probabilidad de fallo en función del tiempo de operación y las condiciones del equipo.

Este proyecto aplica Análisis de Supervivencia y modelos de aprendizaje automático sobre datos de vibración en equipos industriales, con el objetivo de desarrollar una herramienta predictiva robusta. La combinación de técnicas estadísticas tradicionales, como el modelo de Cox Proporcional de Riesgos, con algoritmos modernos de Deep Learning en PyTorch, permitirá mejorar la precisión en la detección de fallos, optimizando la toma de decisiones en el mantenimiento industrial.

# **Descripcion**


El conjunto de datos proviene de sensores instalados en equipos industriales y contiene registros diarios de funcionamiento. Las variables medidas incluyen:

- `Frecuencia_Hz`: frecuencia de vibración del equipo.
- `Aceleracion_RMS_g`: aceleración efectiva (Root Mean Square) en g.
- `Crest_Factor`: indicador de picos en la señal de vibración.
- `Temperatura_C`: temperatura en grados Celsius.
- `Horas_Operacion`: horas acumuladas de operación desde la última intervención.
- `Fallo`: variable binaria (1 si hubo un fallo, 0 si no).
- `Tipo_Fallo`: descripción textual del tipo de fallo (si aplica).

Este dataset se usará para construir modelos que predigan fallos de maquinaria antes de que ocurran, con el fin de optimizar la planificación del mantenimiento.


# **Descripción Dataset**

```python
# Importación de módulos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from lifelines import CoxPHFitter
from sksurv.linear_model import CoxPHSurvivalAnalysis
from sksurv.util import Surv
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, RocCurveDisplay
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import shap

# Cargar el conjunto de datos
datos = pd.read_csv("datos_vibraciones_fallos_tipo.csv")

# Verificación del contenido
print("El conjunto de datos tiene la siguiente forma: {}.".format(datos.shape))
print(datos.head(3))

# Conversión de fecha
datos['Fecha'] = pd.to_datetime(datos['Fecha'])

# Definición de columnas clave
time_column = 'Horas_Operacion'
event_column = 'Fallo'

# One-hot encoding de columnas categóricas
categorias = ['ID_Equipo']
datos_procesados = pd.get_dummies(datos, columns=categorias, drop_first=True)

# Selección de features
features = np.setdiff1d(datos_procesados.columns, [time_column, event_column]).tolist()
print("Columnas definidas como features:")
print(features)

N = len(datos_procesados)
index_train, index_test = train_test_split(range(N), test_size=0.4, random_state=42)
data_train = datos_procesados.loc[index_train].reset_index(drop=True)
data_test = datos_procesados.loc[index_test].reset_index(drop=True)

# Preparación de X, T y E
X_train, X_test = data_train[features], data_test[features]
T_train, T_test = data_train[time_column], data_test[time_column]
E_train, E_test = data_train[event_column], data_test[event_column]

# Mostrar tamaño de cada conjunto
print(f"Tamaño del conjunto de entrenamiento: {len(X_train)} muestras")
print(f"Tamaño del conjunto de prueba: {len(X_test)} muestras")

```

# **Ingenieria de caracteristicas**

```python
datos_procesados['Frecuencia_Log'] = np.log1p(datos_procesados['Frecuencia_Hz'])
datos_procesados['Aceleracion_Cuadrada'] = datos_procesados['Aceleracion_RMS_g'] ** 2
datos_procesados['Temperatura_Factor'] = datos_procesados['Temperatura_C'] * datos_procesados['Crest_Factor']
datos_procesados['Horas_por_Temp'] = datos_procesados['Horas_Operacion'] / (datos_procesados['Temperatura_C'] + 1)

# Revisión rápida
datos_procesados[['Frecuencia_Log', 'Aceleracion_Cuadrada', 'Temperatura_Factor', 'Horas_por_Temp']].head()
```

El análisis de las nuevas variables transformadas revela una estrategia adecuada de **ingeniería de características**, orientada a resaltar patrones ocultos en los datos originales y mejorar la capacidad predictiva del modelo. Por ejemplo, **Frecuencia_Log** aplica una transformación logarítmica a la frecuencia, lo cual ayuda a **reducir la asimetría** y mejorar la linealidad con respecto al objetivo de fallo, facilitando su interpretación en modelos como regresiones o redes neuronales. **Aceleracion_Cuadrada**, al elevar al cuadrado la aceleración RMS, amplifica las diferencias en niveles de vibración, siendo útil para capturar eventos más severos o no lineales. La variable **Temperatura_Factor**, que probablemente combina temperatura con otros indicadores (como crest factor o aceleración), puede representar de forma más integrada el efecto térmico en el desgaste del equipo. Finalmente, **Horas_por_Temp** expresa una razón entre el tiempo de operación y la temperatura, aportando una **medida de eficiencia térmica o esfuerzo térmico acumulado**, útil para estimar envejecimiento acelerado por sobrecalentamiento. Estas nuevas variables son técnicamente justificables, ya que integran el conocimiento físico del sistema y estadísticas robustas, permitiendo al modelo diferenciar más eficazmente entre condiciones normales y estados de falla.

# **Variables predictoras y variable objetivo**

```python
# Separación de variables predictoras y objetivo
X = df_modelo.drop("Fallo", axis=1)
y = df_modelo["Fallo"]

print("Tamaño del set de datos:")
print("X:", X.shape)
print("y:", y.shape)
```

# **Modelo de Clasificación – Random Forest**


En esta sección se entrena un modelo supervisado de clasificación para predecir fallos en función de variables operativas. A pesar del bajo rendimiento del modelo, se identificaron variables clave y se justificó la necesidad de explorar enfoques alternativos como el clustering y análisis de supervivencia.



```python

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Entrenamiento
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train, y_train)

# Predicción
y_pred = modelo_rf.predict(X_test)
y_prob = modelo_rf.predict_proba(X_test)[:, 1]

# Reporte de métricas
print("Matriz de confusión:")
print(confusion_matrix(y_test, y_pred))
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))
print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.4f}")
```

```python
# Curva ROC
RocCurveDisplay.from_estimator(modelo_rf, X_test, y_test)
plt.title("Curva ROC – Random Forest")
plt.show()
```

El bajo valor del AUC (0.42) obtenido con el modelo Random Forest permitió identificar una limitación fundamental: los patrones de fallo no pueden ser capturados adecuadamente mediante técnicas de clasificación tradicionales. Este hallazgo fue clave para reorientar el análisis hacia enfoques más adecuados, como los modelos de supervivencia, que consideran no solo la ocurrencia del fallo sino también el tiempo hasta que sucede.

# **Modelado Predictivo con Análisis de Supervivencia**

El análisis de supervivencia permite modelar el tiempo hasta que ocurre un evento (en este caso, un fallo). Utilizaremos el modelo de Cox Proporcional de Riesgos, ampliamente usado para este tipo de análisis.

Este modelo asume que el efecto de los predictores sobre el riesgo de fallo es proporcional en el tiempo. Los coeficientes del modelo pueden interpretarse como la influencia relativa de cada variable sobre la probabilidad de fallo.


```python

# Creamos un dataframe para supervivencia
df_surv = df_modelo.copy()
df_surv.rename(columns={"Horas_Operacion": "duration", "Fallo": "event"}, inplace=True)

# Confirmamos estructura
df_surv.head()
```

```python
# Entrenamiento del modelo
cox = CoxPHFitter()
cox.fit(df_surv, duration_col='duration', event_col='event')

# Resumen de resultados
cox.print_summary()
```

```python
# Diagrama de efectos (log hazard ratios)
cox.plot()
plt.title("Impacto de las variables sobre el riesgo de fallo (modelo de Cox)")
plt.show()
```

El análisis de regresión de Cox estima el impacto de cada variable sobre el riesgo de fallo, mostrando que tanto Temperatura como Frecuencia podrían estar asociadas a una ligera disminución del riesgo (valores negativos de log(HR)), mientras que Aceleración RMS muestra una mayor variabilidad y potencial asociación con incremento del riesgo. Este enfoque permite cuantificar cómo influye cada variable en la duración operativa antes del fallo.

# **Curvas de supervivencia por tipo de fallo**

```python
from lifelines import KaplanMeierFitter

kmf = KaplanMeierFitter()
for tipo in datos['Tipo_Fallo'].dropna().unique():
    mask = datos['Tipo_Fallo'] == tipo
    kmf.fit(durations=datos[mask][time_column], event_observed=datos[mask][event_column], label=tipo)
    kmf.plot_survival_function()

plt.title("Curvas de supervivencia por tipo de fallo")
plt.xlabel("Horas de operación")
plt.ylabel("Probabilidad de supervivencia")
plt.show()
```

El análisis de las curvas de supervivencia indica que los fallos por **juego mecánico** tienden a ocurrir antes de las **500 horas de operación**, mostrando una rápida disminución en la probabilidad de supervivencia. En cambio, los equipos que presentan **desgaste de rodamientos** suelen mantener una alta probabilidad de funcionamiento hasta aproximadamente las **700–800 horas**, lo que sugiere un proceso de degradación más lento. Por su parte, las fallas por **desbalanceo** y **alineación incorrecta** se distribuyen entre las **500 y 700 horas**, reflejando una severidad y aparición intermedia. Esta información es clave para definir ventanas de mantenimiento y alertas tempranas según el tipo de fallo predominante.

# **Propuesta de automatización**

La modularización del código en funciones reutilizables representa un buen estándar de ingeniería de software en ciencia de dato

```python
def entrenar_modelo_xgb(X, y):
    sm = SMOTE(random_state=42)
    X_res, y_res = sm.fit_resample(X, y)
    X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42)

    modelo = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)
    print(classification_report(y_test, y_pred))
    return modelo
```

Sin embargo, este resultado no fue una limitación sino un hallazgo importante: evidenció que los fallos en equipos rotativos no responden bien a modelos supervisados clásicos, debido a su comportamiento no lineal, la falta de equilibrio en la clase objetivo y la alta dependencia del contexto operativo. Esto justificó la incorporación de técnicas más adecuadas como el análisis de supervivencia, que permitió modelar cuándo ocurre un fallo y no solo si ocurrirá, entregando así una herramienta más efectiva para anticipar fallos con base en el tiempo de operación y condiciones reales del equipo. El enfoque resultó más alineado con las necesidades reales del mantenimiento predictivo, aportando valor al diseño de planes de intervención basados en riesgo.

El análisis permitió segmentar los registros en diferentes perfiles de comportamiento vibracional mediante técnicas de clustering, identificando grupos con mayor propensión al fallo. Además, el modelo de supervivencia tipo Cox permitió asociar variables clave como aceleración RMS, temperatura y frecuencia con el riesgo de fallo, estableciendo relaciones cuantificables para priorizar acciones de mantenimiento. A diferencia del mantenimiento reactivo tradicional, que responde una vez ocurrido el fallo, este enfoque predictivo anticipa eventos críticos, permitiendo reducir los tiempos de inactividad no planificados y optimizar el uso de recursos técnicos y logísticos. Si bien se evidenció que los modelos supervisados por sí solos no son suficientes, el análisis temporal y la segmentación de riesgo demostraron ser herramientas poderosas para disminuir costos operativos y mejorar la confiabilidad de los activos rotativos.
